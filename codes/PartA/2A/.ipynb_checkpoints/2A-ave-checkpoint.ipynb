```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batch_sizes = [4,8,16,32,64]
all_average_metrics = []
```

```python
def plot_loss_and_save(batch_size,fold_average_stats):
    plt.plot(fold_average_stats.epoch,fold_average_stats.loss)
    plt.plot(fold_average_stats.epoch,fold_average_stats.val_loss)
    plt.title('batch_size %s ave model loss'%(batch_size))
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train','validate'], loc= 'upper right' )
    plt.ylim(0,0.4)
    plt.savefig('./figures/2A_batch_%s_average_loss.png'%(batch_size))
    plt.show()
```

```python
def plot_acc_and_save(batch_size,fold_average_stats):
    plt.plot(fold_average_stats.epoch,fold_average_stats.accuracy)
    plt.plot(fold_average_stats.epoch,fold_average_stats.val_accuracy)
    plt.title('batch_size %s ave accuracy'%(batch_size))
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train','validate'], loc= 'lower right' )
    plt.ylim(0.7,1)
    plt.savefig('./figures/2A_batch_%s_average_accuracy.png'%(batch_size))
    plt.show()
```

```python
#Read in results from 2A-overfit and draw the average cross-validation accuracy curve for each hyperparameter models
for batch_size in batch_sizes:
    fold_0_df = pd.read_csv("./no_early/2A_fold_0_batch_%s.csv"%(batch_size))
    fold_1_df = pd.read_csv("./no_early/2A_fold_1_batch_%s.csv"%(batch_size))
    fold_2_df = pd.read_csv("./no_early/2A_fold_2_batch_%s.csv"%(batch_size))
    fold_3_df = pd.read_csv("./no_early/2A_fold_3_batch_%s.csv"%(batch_size))
    fold_4_df = pd.read_csv("./no_early/2A_fold_4_batch_%s.csv"%(batch_size))
    average_metrics = (fold_0_df + fold_1_df + fold_2_df  + fold_3_df  + fold_4_df)/5
    plot_acc_and_save(batch_size,average_metrics)
    plot_loss_and_save(batch_size,average_metrics)
    all_average_metrics.append(average_metrics)
```

```python
#Plot all test accuracy against epoch for comparison
for model_idx in all_average_metrics:
    plt.plot(model_idx.epoch,model_idx.val_accuracy)
plt.title('ave test accuracy vs epoch for different batch_size')
plt.ylabel('test accuracy')
plt.xlabel('epoch')
plt.legend([4,8,16,32,64], loc= 'lower right' )
plt.ylim(0.7,1)
plt.savefig('./figures/2A_test_Acc_tgt.png')
plt.show()
```

```python
#Plot all test accuracy against epoch for comparison
for model_idx in all_average_metrics:
    plt.plot(model_idx.epoch,model_idx.val_loss)
plt.title('ave test loss vs epoch for different batch_size')
plt.ylabel('test loss')
plt.xlabel('epoch')
plt.legend([4,8,16,32,64], loc= 'upper right' )
plt.ylim(0.2,1)
plt.savefig('./figures/2A_test_loss_tgt.png')
plt.show()
```

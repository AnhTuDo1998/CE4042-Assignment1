```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

import os
if not os.path.isdir('proj_figures'):
    os.makedirs('proj_figures')
```

```python
# scaler for data
def scale(X, X_min, X_max):
    return (X - X_min)/(X_max-X_min)
```

```python
NUM_CLASSES = 3

epochs = 1500
batch_size = 32
num_neurons = 10
seed = 10
weight_decay = 1e-6

num_attributes = 21

np.random.seed(seed)
tf.random.set_seed(seed)

histories={}
```

```python
dataset = np.genfromtxt('ctg_data_cleaned.csv', delimiter= ',')
X, Y = dataset[1:,:21], dataset[1:,-1].astype(int)
Y = Y - 1 #For python that start from 0

print(X)

#Split to train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = seed, shuffle=True)

#Scale based on training set 
X_test = scale(X_test,np.min(X_train,axis = 0),np.max(X_train, axis=0))
X_train = scale(X_train, np.min(X_train, axis = 0),np.max(X_train, axis = 0 ))
```

```python
#Model definition
model = keras.Sequential()
model.add(keras.layers.Dense(num_neurons, activation = 'relu',input_dim = num_attributes,kernel_regularizer=keras.regularizers.l2(weight_decay)))
model.add(keras.layers.Dense(NUM_CLASSES, activation ='softmax'))

#default of sgd is lr = 0.01 so no need specify
model.compile(optimizer='sgd',
             loss = keras.losses.SparseCategoricalCrossentropy(),
             metrics=['accuracy'])
```

```python
#Define callback to csv logger
csv_logger = keras.callbacks.CSVLogger("Part1A.csv")
#early_stopper = keras.callbacks.EarlyStopping(monitor='val_loss',patience=20)
#Add early_stopper to the following list for validating the test error convergences
callbacks = [csv_logger]
```

```python
#train until overfitting at 1500 epochs
histories['1A'] =model.fit(X_train, Y_train,validation_data=(X_test,Y_test),
                                        epochs=epochs,
                                        verbose = 2,
                                        batch_size=batch_size,
                                        callbacks=callbacks)
```

```python
#Plot graphs for accuracy and errors
# summarize history for accuracy
plt.plot(histories['1A'].history['accuracy'])
plt.plot(histories['1A'].history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'], loc= 'lower right' )
plt.savefig('./proj_figures/1A_Acc.png')
plt.show()
```

```python
plt.plot(histories['1A'].history['loss'])
plt.plot(histories['1A'].history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'], loc= 'upper right' )
plt.savefig('./proj_figures/1A_loss.png')
plt.ylim(0.2,0.5)
plt.savefig('./proj_figures/1A_loss_minimum.png')
plt.show()
```

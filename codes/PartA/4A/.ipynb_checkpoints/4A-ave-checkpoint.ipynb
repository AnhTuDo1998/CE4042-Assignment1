```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
beta_decay = [0,1e-3,1e-6,1e-9,1e-12]
all_average_metrics = []
```

```python
def plot_loss_and_save(beta,fold_average_stats):
    plt.plot(fold_average_stats.epoch,fold_average_stats.loss)
    plt.plot(fold_average_stats.epoch,fold_average_stats.val_loss)
    plt.title('beta %s ave model loss'%(beta))
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train','validate'], loc= 'upper right' )
    plt.ylim(0,1)
    plt.savefig('./figures/4A_beta_%s_average_loss.png'%(beta))
    plt.show()
```

```python
def plot_acc_and_save(beta,fold_average_stats):
    plt.plot(fold_average_stats.epoch,fold_average_stats.accuracy)
    plt.plot(fold_average_stats.epoch,fold_average_stats.val_accuracy)
    plt.title('beta %s ave accuracy'%(beta))
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train','validate'], loc= 'lower right' )
    plt.ylim(0.7,1)
    plt.savefig('./figures/4A_beta_%s_average_accuracy.png'%(beta))
    plt.show()
```

```python
#Read the results from 4A-overfit and generates the average curve across 5 folds for each hyperparameter instances
for beta in beta_decay:
    fold_0_df = pd.read_csv("./no_early/4A_fold_0_beta_%s.csv"%(beta))
    fold_1_df = pd.read_csv("./no_early/4A_fold_1_beta_%s.csv"%(beta))
    fold_2_df = pd.read_csv("./no_early/4A_fold_2_beta_%s.csv"%(beta))
    fold_3_df = pd.read_csv("./no_early/4A_fold_3_beta_%s.csv"%(beta))
    fold_4_df = pd.read_csv("./no_early/4A_fold_4_beta_%s.csv"%(beta))
    average_metrics = (fold_0_df + fold_1_df + fold_2_df  + fold_3_df  + fold_4_df)/5
    all_average_metrics.append(average_metrics)
    plot_acc_and_save(beta,average_metrics)
    plot_loss_and_save(beta,average_metrics)
```

```python
#Plot all test accuracy against epoch for comparison
for model_idx in all_average_metrics:
    plt.plot(model_idx.epoch,model_idx.val_accuracy)
plt.title('ave test accuracy vs epoch for different beta decay')
plt.ylabel('test accuracy')
plt.xlabel('epoch')
plt.legend([0,1e-3,1e-6,1e-9,1e-12], loc= 'lower right' )
plt.ylim(0.8,0.925)
plt.savefig('./figures/4A_test_Acc_tgt.png')
plt.show()
```

```python
#Plot all test accuracy against epoch for comparison
for model_idx in all_average_metrics:
    plt.plot(model_idx.epoch,model_idx.val_loss)
plt.title('ave test loss vs epoch for different beta decay')
plt.ylabel('test loss')
plt.xlabel('epoch')
plt.legend([0,1e-3,1e-6,1e-9,1e-12], loc= 'upper right' )
plt.ylim(0.2,0.4)
plt.savefig('./figures/4A_test_loss_tgt.png')
plt.show()
```
